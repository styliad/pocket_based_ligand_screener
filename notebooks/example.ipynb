```python
from pathlib import Path

from dock_ligand_annotator.interactions import Interactions
from dock_ligand_annotator.io import (
    load_universe,
    load_protein_mol,
    load_docked_poses
)
```

```python
protein_file = Path("./data/protein/example_protein_7JVQ.pdb")
ligand_file = "./data/ligand/example_ligands.sdf"
```

```python
u = load_universe(protein_file)
protein_mol = load_protein_mol(u)
docked_poses = load_docked_poses(ligand_file)
```

```python
interactions = Interactions(protein_mol, docked_poses)
fp_list = interactions.calculate()
```

```python
fp_list
```

```python
parsed_fps = interactions.parse(fp_list, u)
```

```python
fg_annotated_fps = interactions.annotate(parsed_fps)
```

```python
isinstance(fg_annotated_fps, list)
```

```python
# Save annotated interactions to CSV
interactions.to_csv(fg_annotated_fps, "annotated_interactions.csv")
```
